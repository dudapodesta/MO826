```sparql
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT ?disease ?value
FROM <http://purl.obolibrary.org/obo/merged/DOID>
WHERE {
    ?parent2 rdfs:subClassOf ?parent1.
    ?parent1 rdfs:label "neurodegenerative disease" .
    ?disease rdfs:subClassOf ?parent2.
    ?disease rdfs:label ?value .
}
```

```sparql
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT ?disease ?value
FROM <http://purl.obolibrary.org/obo/merged/DOID>
WHERE {
    ?disease rdfs:subClassOf ?parent1.
    ?parent1 rdfs:label "neurodegenerative disease" .
    ?disease rdfs:label ?value .
}
```
